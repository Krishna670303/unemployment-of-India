```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heat_maps = df_clean[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']]
heat_maps = heat_maps.corr()

#plottingthe figure
plt.figure(figsize=(10,6))
sns.set_context('notebook',font_scale=1)
sns.heatmap(heat_maps, annot=True)
```

## Insights of this char
1 - Unemployment and Employment have a weak negative correlation (-0.22)  as unemployment rises, employment slightly falls.
2 - Unemployment and Labour Participation have almost zero correlation (0.0026)  they do not move together.
3 - Employment and Labour Participation also show no meaningful correlation (0.011)  changes in one dont affect the other.
4 - Overall, the three variables are mostly independent  no strong linear relationships exist inthisdataset.

```python
plt.figure(figsize=(10,4))
sns.barplot(x='Region' , y = 'Estimated Unemployment Rate' , data = df_clean , palette = 'winter')
plt.title("unemployment Rate by Region")
plt.show()
```

Urban unemployment rate is higher than rural

1. From the bar heights, the Urban region has a noticeably higher Estimated Unemployment Rate (around 1314%) compared to the Rural region (around 10%).
This indicates unemployment is more severe in urban areas in your dataset.

2. Error bars show variability or confidence

The vertical black lines (error bars) represent standard deviation or confidence intervals.
Urban has a slightly larger error bar, meaning unemployment rates vary more among the urban samples compared to rural ones.

3. Seaborn barplot() is showing the mean by default

seaborn.barplot() automatically computes and displays the mean of the selected variable (Estimated Unemployment Rate) for each category (Region).
So the bars you see are not raw values, but aggregated averages.

4. The plot uses a custom aesthetic style
You chose:

figsize=(10, 5)  wide and clean chart

palette='winter'  cool color tones



conclusion

The chart highlights that unemployment is noticeably higher in urban regions compared to rural areas, with urban data also showing greater variability. Since the bar heights represent mean values calculated by seaborn.barplot(), the visualization clearly indicates that urban unemployment is both higher and less consistent. The chosen chart style further makes these differences easy to observe.

warnings.filterwarnings("ignore") hides non-critical warnings so the notebook outputstaysclean.

conclusion

The scatterplot reveals a clear relationship between Estimated Employment and Labour Participation, with noticeable differences between Rural and Urban regions. By separating regions through color coding, the chart makes these contrasting patterns easy to interpret. 4. Recovery begins after the lockdown period

After the dip in mid-2020, both rural and urban participation rates start rising again, with rural recovering faster and reaching closer to pre-lockdown levels, while urban recovers more slowly.

conclusion

The labour participation trends show that rural areas consistently maintain higher participation rates than urban regions throughout 20192020. Overall, the chart highlights clear regional differences as well as the strong impact of the pandemic on labour activity.

Unemployment Rate: 0%20% This means most regions have moderate participation and relatively low unemployment.

4. Outliers with extremely high unemployment

There are some clear outlier points where unemployment shoots above 6080%, regardless of participation rate. These indicate specific regions or periods with extreme unemployment conditions.

conclusion

The chart shows that there is no clear linear relationship between labour participation and unemployment, as the points are widely scattered across the plot. Most regions fall within moderate participation levels (4060%) and relatively low unemployment (020%), but several outliers display extremely high unemployment despite varying participation rates. This pattern suggests that higher participation does not necessarily reduce unemployment, and that unemployment levels are influenced by other economic or regional factors beyond participation alone.

```python
states = df_clean['states'].unique()[:5]  # limit to 5 states for visibility

plt.figure(figsize=(12,6))
for s in states:
    data = df_clean[df_clean['states'] == s]
    plt.plot(data['Date'], data['Estimated Unemployment Rate'], label=s)

plt.title("Unemployment Trend for First 5 States")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

insight of this chart

1. Bihar shows the highest unemployment spike Among the five states, Bihar has a sharp peak around 5060%, which is significantly higher than the others. This indicates a period of severe unemployment volatility in Bihar.

2. Delhi also shows noticeable fluctuations Delhis line shows a mid-level spike around 4045%. This suggests Delhi experienced temporary unemployment surges, though not as extreme as Bihar.

3. Andhra Pradesh and Assam remain relatively stable These two states show lower and smoother unemployment trends, without extreme peaks. Their unemployment rates appear more controlled and predictable.

4. Chhattisgarh shows a rising trend Chhattisgarhs line gradually increases over time, indicating a steady upward trend rather than sudden spikes. This might point to long-term unemployment challenges developing in the state.

conclusion

The chart highlights clear differences in unemployment trends across the five states. Bihar stands out with the most severe spike, reaching extremely high unemployment levels, indicating a period of major economic disruption. Delhi also experiences noticeable fluctuations, though less extreme than Bihar, suggesting short-term instability. In contrast, Andhra Pradesh and Assam show relatively stable and lower unemployment patterns, reflecting more steady labour market conditions. Chhattisgarh, however, displays a gradual upward trend, pointing to a slowly developing unemployment challenge over time. Overall, the chart shows a mix of volatility, stability, and emerging long-term concerns across different states.
